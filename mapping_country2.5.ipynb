## compile

```python
%%time 

def Tag_subregion(x):
    
    # OBOR: 6 parts
    if x in NorthEastAsia:
        return 'Northeast Asia'
    elif x in SouthEastAsia:
        return 'Southeast Asia'
    elif x in CentralAsia:
        return 'Central Asia'
    elif x in MiddleEastandNorthAfrica:
        return 'MiddleEast & NorthAfrica'
    elif x in SouthAsia:
        return 'South Asia'
    elif x in MiddleEastEurope:
        return 'MiddleEast Europe'

    # Non-OBOR: 5 continents (4 +  )
    elif x in RestofEurope:
        return 'Rest of Europe'
    elif x in NorthAfrica:
        return 'North Africa'
    elif x in EastAfrica:
        return 'East Africa'
    elif x in WestAfrica:
        return 'West Africa'
    elif x in MiddleAfrica:
        return 'Middle Africa'
    elif x in SouthAfrica:
        return 'South Africa'
    elif x in Australia:
        return 'Australia'
    elif x in NorthAmerica:
        return 'North America'
    elif x in LatinAmerica:
        return 'Latin America'
    else:
        return 'Nowhere'

country_data['subregion'] = country_data['0'].apply(Tag_subregion)
```

```python
country_data['subregion'].value_counts()
```

```python
%%time 

# By Continent
# OBOR: 6 parts
# Non-OBOR: 5 parts (EastAsia, Africa, RestofEurope, Australia, NorthAmerica, LatiAmerica)

flatten = lambda l: [item for sublist in l for item in sublist]

Asia = flatten([NorthEastAsia, SouthEastAsia, CentralAsia, MiddleEastandNorthAfrica, SouthAsia]) + ['Japan', 'North Korea', 'South Korea']


Africa = flatten(['North Africa', 'East Africa', 'West Africa', 'Middle Africa', 'South Africa'])

Europe = flatten(['MiddleEast Europe', 'Rest of Europe'])



# Austrailia， NorthAmerica， LatinAmerica

def Tag_continent(x):
    
    if x in Asia:
        return 'Asia'
    elif x in Europe:
        return 'Europe'
    elif x in Africa:
        return 'Africa'
    elif x in Australia:
        return 'Australia'
    elif x in NorthAmerica:
        return 'North America'
    elif x in LatinAmerica:
        return 'Latin America'
    else:
        return 'Nowhere'
    
country_data['continent'] = country_data['0'].apply(Tag_continent)
```

```python
country_data['continent'].value_counts()
```

```python
%%time 

# By OBOR
# 2 sub-groups

OBOR_all64 = [ 'NorthEastAsia', 'Southeas tAsia' , 'Central Asia', 
              'Central Asia', 'MiddleEastandNorthAfrica', 'SouthAsia', 'MiddleEastEurope']
#itertools.chain(a, b)

def Tag_OBOR(x):
    
    if x in OBOR_all64:
        return 'OBOR'
    else:
        return 'Non-OBOR'

located_data['OBOR'] = located_data['continent'].apply(Tag_OBOR)
```
